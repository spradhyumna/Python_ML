```python
pistachio.replace(to_replace="Kirmizi_Pistachio",value=1,inplace=True)
pistachio.replace(to_replace="Siit_Pistachio",value=2,inplace=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
x=pistachio.drop(['Class','ECCENTRICITY','SHAPEFACTOR_1','SHAPEFACTOR_2'],axis=1)
y=pistachio['Class']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=100)
```

```python
svc = SVC(kernel='linear',random_state=100)
```

```python
svc.fit(x_train,y_train)
```

```python
print(svc.score(x_train,y_train))
print(svc.score(x_test,y_test))
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
plt.figure()
sns.heatmap(cm,annot=True)
plt.xlabel('prdicted')
plt.ylabel('Truth')
```

# Finding Best Parameters

```python
from sklearn.svm import SVC
grid = {
    'C':[0.1,1,10],
    'kernel' : ["poly","rbf"],
    'degree' : [1,3]}
svx=SVC()
```

```python
grid=GridSearchCV(svx,param_grid=grid,cv=5,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(x_test)
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
svc1 = SVC(kernel='rbf',C=10,degree=1,random_state=100)
svc1.fit(x_train,y_train)
```

```python
print(svc1.score(x_train,y_train))
print(svc1.score(x_test,y_test))
```

```python
y_pred_1=svc1.predict(x_test)
```
