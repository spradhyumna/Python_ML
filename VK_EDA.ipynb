# EDA of Run machine Virat Kohli's Century. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
sns.set_palette("pastel")
vk_df=pd.read_csv('https://raw.githubusercontent.com/spradhyumna/Python_ML/main/71%20Centuries%20of%20Virat%20Kohli.csv')
vk_df.rename(columns={'Column1': 'Venue_city'},inplace=True, errors='raise')
vk_df.drop('Unnamed: 14',axis=1,inplace=True)
vk_df.head()
```

# High Score Against Teams

```python
vk_df['year']=pd.to_datetime(vk_df['Date'])
vk_df['year']=vk_df['year'].dt.year 
vk_df['Date']=vk_df['Date'].replace('08-09-2022','09-08-2022')
vk_df.groupby(['Against'],as_index=False)['Score'].max().head()
```

# Centuries over year 

```python
plt.figure(figsize=(15,5))
plt.yticks(size=8)
plt.xticks(size=8)
ax=sns.lineplot(data=vk_df, x="Date", y="Score",marker='o',color='red') 
ax.set_xticklabels( labels=vk_df.Date,rotation=90)
None
```

# Centuries Against Teams

```python
plt.figure(figsize=(15,`6)) 
bx=sns.countplot(x ='Against', data = vk_df)
for p in bx.patches:
        bx.annotate('{}'.format(p.get_height()), (p.get_x()+0.35, p.get_height()+0.15),size=10)
plt.show()
```

Out of 71 Centuries Virat has 15 Centuries against Australia and 13 against Srilanka 

# Centuries While playing Home or Away

```python
plt.figure(figsize=(15,8))
sns.set(font_scale = 1.5)
cx=sns.countplot(x ='Against', data = vk_df,hue='H/A')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
plt.yticks(size=10)
plt.xticks(size=10)
for p in cx.patches:
    cx.annotate(format(p.get_height()), 
                (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), 
                textcoords = 'offset points',size=10)
plt.show()
```

Out of 15 Centuries against Australia  he scored 9 centuries in Away conditions and 6 in home conditions.  

# Virat Kohli's Hundreds year-by-year

```python
plt.figure(figsize=(30,10))
sns.set(font_scale = 1.5)
dx=sns.countplot(x ='year', data = vk_df)
plt.yticks(size=20)
plt.xticks(size=20)
for q in dx.patches:
    dx.annotate(format(q.get_height()), 
                (q.get_x() + q.get_width() / 2, q.get_height()), ha = 'center', va = 'center', xytext = (0, 10),
                textcoords = 'offset points',size=20)
plt.show()
```

In the Year 2016 and 2017 Virat Kohli Scored 11 centuries and he is the secon person after Sachin Tendulkar

In 1998 Sachin Tendulkar scored 12 centuries in all formats combined, which is the record for Most centuries in the calender year in cricket in all formats.

# Whether Virat is Out or Not Out when he hit Century

```python
plt.figure(figsize=(30,10))

sns.set(font_scale = 1.5)
wx=sns.countplot(x ='Against', data = vk_df,hue='Out/Not Out')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
plt.yticks(size=20)
plt.xticks(size=20)
for q in wx.patches:
    wx.annotate(format(q.get_height()), 
                (q.get_x() + q.get_width() / 2, q.get_height()), ha = 'center', 
                va = 'center', xytext = (0, 10), textcoords = 'offset points',size=20)
plt.show()
```

```python
plt.figure(figsize=(8,8))
colors = sns.color_palette('pastel')
purpose_freq = vk_df['Out/Not Out'].value_counts()
plt.pie(purpose_freq, autopct='%.2f%%',explode = (0.03,0.03),colors=colors ,labels = purpose_freq.index)

centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
None
```

Out of 71 Centuries, Kohli outs 51 times

He outs mostly for Australia i.e 13 times
